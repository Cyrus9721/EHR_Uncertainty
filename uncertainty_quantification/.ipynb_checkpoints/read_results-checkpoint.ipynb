```python
def read_results(results1):
    df1 = pd.DataFrame(results1).T
    new_df = pd.DataFrame()
    for i in df1.index:
        temp_df = pd.DataFrame.from_dict(df1.loc[i, 0] , orient='index', columns=['Value']).T
        new_df = pd.concat([new_df, temp_df], axis=0)
    new_df.index = df1.index
    new_df = pd.concat([df1, new_df], axis = 1)
    new_df = new_df.drop(columns=0)
    return new_df.round(4).loc[:,  ['test/cls/Acc', 'test/cls/Brier', 'test/cls/NLL','test/cal/ECE','test/cal/aECE']]
```

```python
def read_results(results1):
    df1 = pd.DataFrame(results1).T
    new_df = pd.DataFrame()
    for i in df1.index:
        temp_df = pd.DataFrame.from_dict(df1.loc[i, 0] , orient='index', columns=['Value']).T
        new_df = pd.concat([new_df, temp_df], axis=0)
    new_df.index = df1.index
    new_df = pd.concat([df1, new_df], axis = 1)
    new_df = new_df.drop(columns=0)
    return new_df.round(4).loc[:,  ['test/cls/Brier']]
```

```python
def format_latex_with_color(df1, df2):
    latex_str = ""
    # Ensure both DataFrames are the same size
    if df1.shape != df2.shape:
        raise ValueError("DataFrames must be of the same size")
    
    # Iterate through each row and column index
    l1 = ''
    l2 = ''
    for i in range(df1.shape[0]):
        row_str1 = "& "
        row_str2 = "& "
        for j in range(df1.shape[1]):
            val1 = df1.iloc[i, j]
            val2 = df2.iloc[i, j]
            # Apply color to the smaller value
            if val1 < val2:
                row_str1 += f"\\cellcolor{{gray!12}}{val1:.4f} & "
                row_str2 += f"{val2:.4f} & "
            elif val1 > val2:
                row_str1 += f"{val1:.4f} & "
                row_str2 += f"\\cellcolor{{gray!12}}{val2:.4f} & "
            else:  # values are equal, no color applied
                row_str1 += f"{val1:.4f} & "
                row_str2 += f"{val2:.4f} & "
        
        # Remove the last '&'
        row_str1 = '& ' + row_str1.strip('& ') + " \\\\\n"
        row_str2 = '& ' + row_str2.strip('& ') + " \\\\\n"
        # Add the rows to the main string
        l1 += row_str1
        l2 += row_str2
        latex_str += row_str1 + row_str2
    
    return l1 + l2
```

```python
def df_to_latex(df):
    latex_str = ""
    for index, row in df.iterrows():
        # Create a string for each row and append it to the main string
        row_str = " & " + " & ".join([f"{val:.4f}" for val in row]) + " \\\\\n"
        latex_str += row_str
    return latex_str
```
