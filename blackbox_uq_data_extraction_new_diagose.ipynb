```python
df_new_hypertension = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_hypertension/labeled_patients.csv')
df_new_hyperlipidemia = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_hyperlipidemia/labeled_patients.csv')
# df_new_pancan = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_pancan/labeled_patients.csv')
# df_new_celiac = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_celiac/labeled_patients.csv')
# df_new_lupus = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_lupus/labeled_patients.csv')
df_new_acutemi = pd.read_csv('EHRSHOT_ASSETS/benchmark/new_acutemi/labeled_patients.csv')

df_new_hypertension = df_new_hypertension.drop(columns=['label_type'])
df_new_hyperlipidemia = df_new_hyperlipidemia.drop(columns=['label_type'])
# df_new_pancan = df_new_pancan.drop(columns=['label_type'])
# df_new_celiac = df_new_celiac.drop(columns=['label_type'])
# df_new_lupus = df_new_lupus.drop(columns=['label_type'])
df_new_acutemi = df_new_acutemi.drop(columns=['label_type'])

df_new_hypertension = df_new_hypertension.rename(columns={'value': 'value_new_hypertension'})
df_new_hyperlipidemia = df_new_hyperlipidemia.rename(columns={'value': 'value_new_hyperlipidemia'})
# df_new_pancan = df_new_pancan.rename(columns={'value': 'value_new_pancan'})
# df_new_celiac = df_new_celiac.rename(columns={'value': 'value_new_celiac'})
# df_new_lupus = df_new_lupus.rename(columns={'value': 'value_new_lupus'})
df_new_acutemi = df_new_acutemi.rename(columns={'value': 'value_new_acutemi'})

df_merged = pd.merge(df_new_hypertension, df_new_hyperlipidemia, on=['prediction_time', 'patient_id'], how='inner')
# df_merged = pd.merge(df_merged, df_new_pancan, on=['prediction_time', 'patient_id'], how='inner')
# df_merged = pd.merge(df_merged, df_new_celiac, on=['prediction_time', 'patient_id'], how='inner')
# df_merged = pd.merge(df_merged, df_new_lupus, on=['prediction_time', 'patient_id'], how='inner')
df_merged = pd.merge(df_merged, df_new_acutemi, on=['prediction_time', 'patient_id'], how='inner')
```

```python
df_merged.to_csv('white_box_common_data/new_diagnose_blood.csv', index=False)
```
